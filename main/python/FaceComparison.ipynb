```python
def compare_faces(image1, image2):
  if image1 is None or image2 is None:
    return np.inf
  image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
  image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
  face_features1 = face_cascade.detectMultiScale(image1_gray, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40))
  face_features2 = face_cascade.detectMultiScale(image2_gray, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40))
  if len(face_features1) == 0 or len(face_features2) == 0:
    return np.inf
  distance = np.linalg.norm(face_features1[0] - face_features2[0])
  return distance
```

```python
is_matched = False
for dataset_image in dataset_images:
  distance = compare_faces(image, dataset_image)

  # If the distance between two faces is below a certain threshold, then the faces are considered to be a match
  if distance < 0.5:
    is_matched = True
    break

# Print a message to indicate whether the uploaded image was matched with any image in the dataset
if is_matched:
  print("The uploaded image was matched with an image in the dataset.")
else:
  print("The uploaded image was not matched with any image in the dataset.")
```
